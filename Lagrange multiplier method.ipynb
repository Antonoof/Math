```python
from sympy import *
```

```python
x, y, l = symbols('x y l')
f1 = x**2 + y**2 - 4*x - 2*y + 5 + l*(x + y - 3)
```

```python
Derivative(f1, x)
```

```python
diff_x = diff(f1, x)

print(diff_x)
```

```python
def Lagrange(f):
  """
  F строго зависит от x, y, l
  """
  x, y, l = symbols('x y l')

  # Находим частные производные
  diff_x = diff(f, x)
  diff_y = diff(f, y)
  diff_l = diff(f, l)
  print(f'dL/dx = {diff_x}\ndL/dy = {diff_y}\ndL/dl = {diff_l}\n')

  # Решение СЛАУ
  res = linsolve([diff_x, diff_y, diff_l], (x, y, l))
  return res
```

```python
Lagrange(f1)
```
